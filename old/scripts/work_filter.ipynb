```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(1,10,100)
y=np.sin(x)
plt.plot(x,y)
```

```python

window=np.hamming(len(x))
result=window*y
plt.plot(x,result,label="result")
plt.plot(x,window,label="window")
plt.plot(x,y,label="input")
plt.legend()
window.sum()
```
